# disaggregating the tables

## games table

```python
games = df_joined.get(["Games", "Year", "Season", "City"]).copy()
games.head()
```

*Check if the games take place in more than one city*

```python
games = games.sort_values(by = ["Year", "Season"])
games = games.drop_duplicates(["Games"])
games["id_games"] = range(len(games))
games = games.reset_index(drop = True)
games = games[["id_games"] + [column for column in games.columns if column != "id_games"]]
```

Side note: please don't use `inplace=True`

```python
games.head()
```

```python
games.tail()
```

```python
games.info()
```

```python
games.to_csv("data/olympics/games.csv", index = False, encoding = 'latin1')
```

## events

```python
events = df_joined.get(["Games", "Sport", "Event"])
events = events.drop_duplicates()
events.head()
```

```python
len(events)
```

```python
ids_games = games.get(["id_games", "Games"])
```

```python
events = pandas.merge(events, ids_games, on = "Games", how = "left")
```

```python
len(events)
```

```python
events = events.sort_values(by = "Games")
events["id_event"] = range(len(events))
events = events[["id_event", "Games", "Sport", "Event", "id_games"]]
```

```python
events
```

```python
events.to_csv("data/olympics/events.csv", index = False, encoding = "latin1")
```

## medalist info

```python
medalists = df_joined.loc[~df_joined["Medal"].isnull()].copy()
medalists = medalists.get(["Games", "Year", "Season", "NOC", "ID", "Name", "Event", "Sport", "Medal"])
medalists = medalists.rename(columns = {"ID": "id_athlete"})
medalists = medalists.sort_values(by = ["Year", "Season", "NOC"])
medalists = medalists.reset_index(drop = True)
medalists.head()
```

```python
len(medalists)
```

```python
medalists = pandas.merge(medalists, events[["id_event", "Games", "Sport", "Event"]], on =["Games", "Sport", "Event"])
```

```python
medalists = medalists[["id_athlete", "id_event", "Medal"]]
```

```python
medalists
```

```python
medalists = medalists.rename(columns = {"id_athlete": "athlete_id"})
medalists.head()
```

```python
medalists.to_csv("data/olympics/medalist.csv", index = False)
```
