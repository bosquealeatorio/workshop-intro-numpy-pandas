# Pokedex example

Pandas helps with the scrapping of tables in html format. 

* `pandas.read_html` https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.read_html.html
* Pokedex database: https://pokemondb.net/pokedex/all

```python
url = "https://pokemondb.net/pokedex/all"
html_text = requests.get(url).content
```

```python
df_pokedex = pandas.read_html(html_text)[0]
df_pokedex
```

# Some examples

```python
df_pokedex.head()
```

*Electric pokemons*

```python
df_pokedex.loc[df_pokedex["Type"] == "Electric"].head()
```

*Electric pokemons first generation* 

```python
df_pokedex.loc[(df_pokedex["Type"] == "Electric") & 
                (df_pokedex["#"] <= 151)]
```

*Dual type electric pokemon*

```python
df_pokedex.loc[(df_pokedex["Type"].str.contains("Electric")) & 
               (df_pokedex["Type"].str.contains(" "))].head()
```

*Strongest pokemon*

```python
df_pokedex.sort_values(by = "Total", ascending = False)
```

*Most popular type of pokemon*

```python
split_columns = df_pokedex["Type"].str.split(" ", expand = True)

df_pokedex["Type 1"] = split_columns[0]
df_pokedex["Type 2"] = split_columns[1]

df_pokedex.head()
```

```python
df_pokedex.get(["Name", "Type 1"])\
            .groupby("Type 1").count()\
            .sort_values("Name", ascending = False)
```

*Easy graphics*

```python
df_pokedex.get(["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"])\
            .loc[df_pokedex["Type 1"] == "Grass"].boxplot()


_ = plt.title("Stats for main type grass pokemon")
```
